#**Machine Learning**


##Model Building


```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import xgboost as xgb
```

```python
X=data.drop(columns=['loan_amount','Id'])
y=data['loan_amount']
```

```python
X_encoded=pd.get_dummies(X,drop_first=True)
X_encoded
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_encoded,y,test_size=0.2,random_state=42)
```



>**1. Importing train_test_split:**



from sklearn.model_selection import train_test_split imports the train_test_split function, which is specifically designed for splitting data into training and testing sets in machine learning.


>**2. Splitting the Data:**



*X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)*

*   **" X_encoded and y "** represent your entire dataset, where X_encoded likely contains the features (independent variables) and y contains the target variable (dependent variable).

*   **" test_size=0.2 "**
specifies that 20% (0.2) of the data will be allocated to the testing set. The remaining 80% will be used for training.

*   **" random_state=42 "** sets a seed for the random number generator used to shuffle the data before splitting. This ensures reproducibility - if you run the code again with the same random state, you'll get the same split.

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

In essence, the above code snippet prepares your training and testing data for machine learning by ensuring that all features have a similar scale, which can lead to better model performance.

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR

models = {
    'Linear Regression':LinearRegression(),
    'Random Forest':RandomForestRegressor(random_state=42),
    'Gradient Boosting Regressor':GradientBoostingRegressor(random_state=42),
    'Support Vector Regressor':SVR(),
}
```

```python
results={}
for name, model in models.items():
    model.fit(X_train_scaled if name in ['Support Vector Regressor'] else X_train, y_train)
    y_pred=model.predict(X_test_scaled if name in ['Support Vector Regressor '] else X_test)
    mse=mean_squared_error(y_test,y_pred)
    r2=r2_score(y_test,y_pred)
   # core(y_test,y_pred)
    results[name]={'MSE':mse, 'R2':r2}
```

*   **Loops through models:** It iterates over a dictionary (models) containing
different models.

*  **Conditional fitting:** Each model is trained (fit) on the training data (X_train, y_train). For Support Vector Regression (name is "Support Vector Regressor"), the data is scaled (X_train_scaled) before fitting.

*  **Prediction and evaluation:**
Predictions are made on the testing data (X_test). Again, for Support Vector Regression, scaled data (X_test_scaled) is used.
Mean Squared Error (MSE) and R-squared (R²) are calculated to measure prediction accuracy.

*   **Storing results:** The MSE and R² for each model are stored in a dictionary (results) with the model name as the key.
In short, it tests different models (scaling for Support Vector Regression) and stores their performance (MSE, R²) for comparison.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

def preprocess_new_data(new_data, feature_columns, scaler=None):
    new_data_df= pd.DataFrame([new_data])
    new_data_encoded=pd.get_dummies(new_data_df)
    for col in feature_columns:
        if col not in new_data_encoded.columns:
            new_data_encoded[col]=0

    new_data_encoded=new_data_encoded[feature_columns]

    if scaler:
        new_data_scaled=scaler.transform(new_data_encoded)
        return new_data_scaled

    return new_data_encoded

#Function to predict the loan amount
def predict_loan_amount(model, new_customer_data, feature_columns, scaler=None):
  new_data_preprocessed=preprocess_new_data(new_customer_data, feature_columns, scaler)
  # Fit the model if it's not already fitted
  if not hasattr(model, 'coef_'):  # Check if model has been fitted
    model.fit(X_train_scaled if isinstance(model, SVR) else X_train, y_train) # Fit using appropriate data based on model type
  predicted_loan_amount=model.predict(new_data_preprocessed)
  return predicted_loan_amount[0]
```

 Here's a breakdown of the code in a shorter way:

**preprocess_new_data function:**

*   Takes new data, feature columns, and an optional scaler as input.
*   Converts the data to a DataFrame with one row.
*   Applies one-hot encoding to categorical features (if any).
*   Fills missing features (not in new_data) with zeros.
*   Selects only the required features (feature_columns).
*   If a scaler is provided, scales the encoded data and returns it.
*   Otherwise, returns the encoded data without scaling.
*   In essence, this function prepares new data for prediction by encoding *   *   features, handling missing values, and optionally scaling.

**predict_loan_amount function:**

*   Takes a model, new customer data, feature columns, and an optional scaler as input.
*   Preprocesses the new data using the preprocess_new_data function.
*   Checks if the model is already fitted. If not, it fits the model on the appropriate training data (scaled for SVR).
*   Makes a prediction on the preprocessed new data.
*   Returns the predicted loan amount (the first element of the prediction array).
*   In short, this function preprocesses new data and uses the trained model to predict the loan amount.

***These functions improve code reusability by encapsulating data preprocessing and prediction logic.***










```python
new_customer={
    'city':'rural_city',
    'age':35,
    'sex':'male',
    'social_class':'middle',
    'primary_business':'farming',
    'secondary_buysiness':'none',
    'annual_income':50000,
    'monthly_expenses':3000,
    'old_dependents':2,
    'young_dependents':1,
    'home_ownership':'owned',
    'type_of_house':'pucca',
    'occupants_count':4,
    'house_area':1200,
    'sanitary_avaiability':'yes',
    'water_avaiability':'yes',
    'loan_purpose':'agriculture',
    'loan_tenure':24,
    'loan_installments':12

}
trained_model=models['Linear Regression']
feature_columns=X_encoded.columns.tolist()
scaler=None
predicted_amount = predict_loan_amount(trained_model, new_customer, feature_columns, scaler)
```
